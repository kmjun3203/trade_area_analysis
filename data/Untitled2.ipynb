```python
filled = pd.read_csv("./서울시_상권_점포_지출유형_추가.csv", encoding="cp949")

# 공백 -> NaN 정리
for c in ["지출_종류_1","지출_종류_2","지출_종류_3"]:
    filled[c] = filled[c].replace("", pd.NA)

# ✅ 허용값 외 값이 들어가면 바로 에러 내게 해서 실수 방지
allowed_set = set(allowed)
bad = filled.melt(id_vars=["서비스_업종_코드_명"], value_vars=["지출_종류_1","지출_종류_2","지출_종류_3"]) \
            .dropna()
bad = bad[~bad["value"].isin(allowed_set)]
if len(bad) > 0:
    raise ValueError(f"허용되지 않은 지출 종류가 들어있음:\n{bad.head(20)}")

# dict 만들기: 업종 -> [종류1, 종류2, 종류3]
manual_mapping2 = {}
for _, r in filled.iterrows():
    vals = [r["지출_종류_1"], r["지출_종류_2"], r["지출_종류_3"]]
    vals = [v for v in vals if pd.notna(v)]
    if len(vals) > 0:
        manual_mapping2[r["서비스_업종_코드_명"]] = vals[:3]

# 적용
def apply_manual(row):
    name = row["서비스_업종_코드_명"]
    if pd.isna(row["지출_종류_1"]) and name in manual_mapping2:
        vals = manual_mapping2[name]
        row["지출_종류_1"] = vals[0] if len(vals) > 0 else pd.NA
        row["지출_종류_2"] = vals[1] if len(vals) > 1 else pd.NA
        row["지출_종류_3"] = vals[2] if len(vals) > 2 else pd.NA
    return row

df = df.apply(apply_manual, axis=1)

# 결과 확인
df[df["지출_종류_1"].isnull()]["서비스_업종_코드_명"].unique()
```

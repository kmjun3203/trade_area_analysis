```python
import pandas as pd
import numpy as np
```

```python
df_area = pd.read_csv('./서울시_상권분석서비스_좌표변환.csv')
df_road_population = pd.read_csv('./서울시 상권분석서비스(길단위인구-상권).csv', encoding='cp949')
df_resident_population = pd.read_csv('./서울시 상권분석서비스(상주인구-상권).csv', encoding='cp949')
df_worker_population = pd.read_csv('./서울시 상권분석서비스(직장인구-상권).csv', encoding='cp949')
df_store = pd.read_csv('./서울시 상권분석서비스(점포-상권).csv', encoding='cp949')
df_sales = pd.read_csv('./서울시 상권분석서비스(추정매출-상권).csv', encoding='cp949')
df_change = pd.read_csv('./서울시 상권분석서비스(상권변화지표-상권).csv', encoding='cp949')
df_income_consumption = pd.read_csv('./서울시 상권분석서비스(소득소비-상권).csv', encoding='cp949')
df_attraction_facility = pd.read_csv('./서울시 상권분석서비스(집객시설-상권).csv', encoding='cp949')
df_area2 = pd.read_csv('./서울시 상권분석서비스(영역-상권).csv', encoding='cp949')
```

```python
dfs = [
    df_area,
    df_road_population,
    df_resident_population,
    df_worker_population,
    df_store,
    df_sales,
    df_change,
    df_income_consumption,
    df_attraction_facility,
    df_area2,
]

common_names = set(dfs[0]['상권_코드_명'])  # 첫 DF 기준으로 시작

for df in dfs[1:]:
    common_names &= set(df['상권_코드_명'])

len(common_names)  # 교집합 개수 확인
```

```python
df_area2['상권_코드_명'].nunique()
```

```python
df_area               = df_area[df_area['상권_코드_명'].isin(common_names)].copy()
df_road_population    = df_road_population[df_road_population['상권_코드_명'].isin(common_names)].copy()
df_resident_population= df_resident_population[df_resident_population['상권_코드_명'].isin(common_names)].copy()
df_worker_population  = df_worker_population[df_worker_population['상권_코드_명'].isin(common_names)].copy()
df_store              = df_store[df_store['상권_코드_명'].isin(common_names)].copy()
df_sales              = df_sales[df_sales['상권_코드_명'].isin(common_names)].copy()
df_change             = df_change[df_change['상권_코드_명'].isin(common_names)].copy()
df_income_consumption = df_income_consumption[df_income_consumption['상권_코드_명'].isin(common_names)].copy()
df_attraction_facility= df_attraction_facility[df_attraction_facility['상권_코드_명'].isin(common_names)].copy()
df_area2               = df_area2[df_area2['상권_코드_명'].isin(common_names)].copy()
```

```python
df_area.to_csv('./서울시_상권분석서비스_좌표변환_filtered.csv', index=False, encoding='cp949')
df_road_population.to_csv('./서울시 상권분석서비스(길단위인구-상권)_filtered.csv', index=False, encoding='cp949')
df_resident_population.to_csv('./서울시 상권분석서비스(상주인구-상권)_filtered.csv', index=False, encoding='cp949')
df_worker_population.to_csv('./서울시 상권분석서비스(직장인구-상권)_filtered.csv', index=False, encoding='cp949')
df_store.to_csv('./서울시 상권분석서비스(점포-상권)_filtered.csv', index=False, encoding='cp949')
df_sales.to_csv('./서울시 상권분석서비스(추정매출-상권)_filtered.csv', index=False, encoding='cp949')
df_change.to_csv('./서울시 상권분석서비스(상권변화지표-상권)_filtered.csv', index=False, encoding='cp949')
df_income_consumption.to_csv('./서울시 상권분석서비스(소득소비-상권)_filtered.csv', index=False, encoding='cp949')
df_attraction_facility.to_csv('./서울시 상권분석서비스(집객시설-상권)_filtered.csv', index=False, encoding='cp949')
```

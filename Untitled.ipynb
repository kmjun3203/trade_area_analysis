```python
import pandas as pd
from pyproj import Transformer
import folium
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
```

```python
df_area = pd.read_csv("./data/서울시_상권분석서비스_좌표변환.csv")
df_area
```

```python
# 서울의 중심 좌표
seoul_center = [37.5665, 126.9780]

# 지도 생성
m = folium.Map(
    location=seoul_center,
    zoom_start=11,
    tiles='OpenStreetMap'
)

# 마커 클러스터 생성 (마커가 많을 때 자동으로 그룹화)
marker_cluster = MarkerCluster().add_to(m)

# 각 상권 위치에 마커 추가
for idx, row in df_area.iterrows():
    # 팝업 내용 구성
    popup_text = f"""
    <b>{row['상권_코드_명']}</b><br>
    상권구분: {row['상권_구분_코드_명']}<br>
    자치구: {row['자치구_코드_명']}<br>
    행정동: {row['행정동_코드_명']}<br>
    면적: {row['영역_면적']:,}㎡<br>
    좌표: ({row['위도']:.6f}, {row['경도']:.6f})
    """
    
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=folium.Popup(popup_text, max_width=300),
        tooltip=row['상권_코드_명']
    ).add_to(marker_cluster)

# 주피터 노트북에서 지도 표시
m
```

```python
count_by_type = df_area["상권_구분_코드_명"].value_counts()

plt.figure(figsize=(6,4))
count_by_type.plot(kind="bar")
plt.ylabel("상권 수")
plt.title("상권구분별 상권 개수")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
